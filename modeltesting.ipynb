```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Separate features and target variables
colsused = ['Crm Cd', 'Vict Age', 'Vict Sex', 'Vict Descent', 'theft', 'vehicular', 'LAT', 'LON']

X = sparse_df[colsused]
y = sparse_df[['crime_type', 'AREA', 'Weapon Used Cd']]  # Multiple target variables

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define classifiers
classifiers = {
    "Random Forest Classifier": RandomForestClassifier(),
    "KNN Classifier": KNeighborsClassifier(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Gradient Boosting Classifier": GradientBoostingClassifier(),
    "SVC": SVC()
}

# Store accuracy scores for each model
accuracy_scores = {}

# Train and evaluate each classifier for each target variable
for name, clf in classifiers.items():
    clf_accuracy = []
    for target_col in y.columns:
        y_train_target = y_train[target_col]
        y_test_target = y_test[target_col]
        clf.fit(X_train, y_train_target)
        y_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_test_target, y_pred)
        clf_accuracy.append(accuracy)
    accuracy_scores[name] = sum(clf_accuracy) / len(clf_accuracy)

# Print single accuracy score for each model
print("Single accuracy score for each model:")
for name, accuracy in accuracy_scores.items():
    print(f"{name}: {accuracy}")
```

```python
# Store actual vs predicted values for each model
actual_vs_predicted = {}

# Train and evaluate each classifier for each target variable
for name, clf in classifiers.items():
    clf_predictions = {}
    for target_col in y.columns:
        y_train_target = y_train[target_col]
        y_test_target = y_test[target_col]
        clf.fit(X_train, y_train_target)
        y_pred = clf.predict(X_test)
        clf_predictions[target_col] = {'Actual': y_test_target, 'Predicted': y_pred}
    actual_vs_predicted[name] = clf_predictions

# Print actual vs predicted values for each model
print("Actual vs Predicted values for each model:")
for name, predictions in actual_vs_predicted.items():
    print(f"{name}:")
    for target_col, values in predictions.items():
        print(f"Target Variable: {target_col}")
        print("Actual Values:", values['Actual'])
        print("Predicted Values:", values['Predicted'])
        print()
```
